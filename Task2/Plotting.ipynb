```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.DataFrame([(1, '48.055090'), (2, '23.030414'), (3, '15.642484'), (4, '11.878623'), (5, '10.251628'), (6, '8.322426'), (7, '7.888524'), (8, '6.671789'), (10, '5.557927'), (12, '5.071201'), (14, '4.442201'), (15, '4.353948'), (18, '4.082764'), (21, '3.938421'), (24, '3.412710'), (28, '3.705615')])
#df.rename(columns={0: 'Processes', 1: 'Raw_data'}, inplace=True)
#df['Value'], df['Time'] = df['Raw_data'].str.split(' ', 1).str
df.rename(columns={0: 'Processes', 1: 'Time'}, inplace=True)
df['Time'] = pd.to_numeric(df['Time'])
df['Acceleration'] = df['Time'][0] / df['Time']
print(df)
```

```python
plt.figure(figsize=(14, 10))
plt.xlabel('Processes', fontsize=20)
plt.ylabel('Acceleration', fontsize=20)
plt.grid(True)
plt.title('', fontsize=20)
plt.plot(df['Processes'], df['Acceleration'], 'ob-', label='Acceleration N=1000000 T=0.000000001')
plt.legend(fontsize=20, loc='best')
plt.savefig('Acceleration.jpg')
plt.show()
```
