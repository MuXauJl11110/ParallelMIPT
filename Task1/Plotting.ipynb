```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.DataFrame([(1, '3.141593 31.760000'), (2, '3.141593 19.400000'), (3, '3.141593 12.950000'), (4, '3.141593 9.710000'), (5, '3.141593 7.800000'), (6, '3.141593 6.670000'), (7, '3.141593 5.040000'), (8, '3.141593 5.070000'), (10, '3.141593 4.130000'), (12, '3.141593 3.290000'), (14, '3.141593 2.930000'), (15, '3.141593 2.820000'), (18, '3.141593 2.910000'), (21, '3.141593 2.020000'), (24, '3.141593 1.460000'), (28, '3.141593 1.640000')])
df.rename(columns={0: 'Processes', 1: 'Raw_data'}, inplace=True)
df['Value'], df['Time'] = df['Raw_data'].str.split(' ', 1).str
df['Time'] = pd.to_numeric(df['Processes', 'Time'])
df['Acceleration'] = df['Time'][0] / df['Time']
print(df)
```

```python
plt.figure(figsize=(14, 10))
plt.xlabel('Processes', fontsize=20)
plt.ylabel('Acceleration', fontsize=20)
plt.grid(True)
plt.title('', fontsize=20)
plt.plot(df['Processes'], df['Acceleration'], 'ob-', label='Acceleration')
plt.legend(fontsize=20, loc='best')
plt.savefig('Acceleration.jpg')
plt.show()
```
